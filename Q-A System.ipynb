## Scoring Algorithm

Calculating the score of a candidate sentence based on two dimensions:
    
    1) The number of matched bigrams between a question and a candidate sentence
    2) The TD-IDF score between a question and candidate sentence
    
The score of the two dimensions are added together to create the overall score for each candidate.

```python
def score_algo(question,answer):
    # Bigrams
    question_lower = question.lower()
    answer_lower = answer.lower()
    question_lower_words = question_lower.split(' ')
    answer_lower_words = answer_lower.split(' ')
    
    # NEED TO REMOVE STOP WORDS FROM QUESTION
    question_words = question.split(' ')
    answer_words = answer.split(' ')
    bigrams_question = set([b for b in zip(question_words[:-1], question_words[1:])])
    bigrams_answer = set([b for b in zip(answer_words[:-1], answer_words[1:])])
    match_bigrams = bigrams_question.intersection(bigrams_answer)
    bigram_score =  len(match_bigrams)
    
    # TD-IDF Matched
    common_words = set(question_lower_words).intersection(answer_lower_words);
    #Find column indexes in TFIDF matrix:
    common_words = list(common_words)
    col_index = []
    # ONLY CURRENTLY WORKS FOR LOWERCASE
    for i in common_words: col_index.append(tf.vocabulary_[i])
    if answer in row_index:
        row = row_index[answer]
    else:
        return "Error: Candidate answer not in CEO set"
    score_matched = []
    for i in col_index: score_matched.append(tfidf_mat[row,i])
    tfidf_matched = sum(score_matched)
    
    # TF-IDF Non-Matched (Missing)
    missing_words = np.setdiff1d(question_lower_words,answer_lower_words,assume_unique=True)
    tfidf_question = tf.fit_transform(question_lower_words)
    missing_index = []
    for i in missing_words: missing_index.append(tf.vocabulary_[i])
    score_missing = []
    for i in missing_index: score_missing.append(tfidf_question[0,i])
    tfidf_missing = sum(score_missing)
    
    # Overall Score
    score = bigram_score + tfidf_matched - tfidf_missing;
    return score
```

```python
def question_preprocess(question):
    if question[-1] == '?':
        question = question[0:len(question)-1]
    
    tokenized_sent = word_tokenize(question)
    question = ' '.join([word for word in tokenized_sent if word.lower() not in stop_words])
    
    return question
```
